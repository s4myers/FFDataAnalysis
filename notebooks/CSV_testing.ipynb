```python
### Possible way to pull data from csv file ###
yards = 0
td = 0
name = "Marshawn Lynch"
year = "2014"

with open("test.csv") as testfile:
    names =["Year","Player","Week","Yards","TD"]
    years =["2014,2015"]
    reader2 = csv.DictReader(testfile,fieldnames = names)
    for row in reader2:
        if row["Player"]==name and row["Year"]==year:
            yards+=int(row["Yards"])
            td+=int(row["TD"])
print "%s had %s total yards and %s touchdowns in %s" % (name,yards,td,year)   
```
