```python
import csv
import os
main_dir = "Users/sean/Documents/GitHub/FFDataAnalysis/CSV_data"
```

```python
def points_allowed(team,pos,year, 
                   ptd = 4.0,rtd = 6.0,
                   ryard = .1, pyard = 0.04,
                   ppr = 1.0,
                   tover = -2.0):
    """
    This fuction generates the fantasy points allowed for a given year.

    Keyword arguments:
    team  -- the nfl team name
    pos   -- the position that the team allowed points to
    year  -- the year 
    rtd   -- points per rush touchdown allowed
    ptd   -- points per pass touchdown allowed
    ryard -- points per rush yard allowed
    pyard -- points per pass allowed
    ppr   -- points per reception allowed
    tover -- points lost by player due to a turnover
    """
    
    # Initialize variables
    tot_ryards = 0.0
    tot_pyards = 0.0
    tot_rec = 0
    tot_rtd = 0
    tot_ptd = 0
    tot_tover = 0
    

    # setup of paths
    csv_file_name = pos +"Stats.csv"
    csv_path = os.path.join("c:/",main_dir,csv_file_name)
    
    # read from the csv file and get total points allowed to a QB
    with open(csv_path) as csv_file:
        reader = csv.DictReader(csv_file)
        for row in reader:
            if (row["Opp"]==team or row["Opp"]=='@'+team) and row["Year"]==year and row["GS"]=="1":
                tot_pyards += 0.0 if row["PYds"]=="--" else float(row["PYds"])
                tot_ryards += 0.0 if row["RYds"]=="--" else float(row["RYds"])
                tot_ptd += 0.0 if row["PTD"]=="--" else float(row["PTD"])
                tot_rtd += 0.0 if row["RTD"]=="--" else float(row["RTD"])
                tot_tover += 0.0 if row["Lost"]=="--" else int(row["Lost"])
                tot_tover += 0.0 if row["Int"]=="--" else int(row["Int"])
    tot_points = tot_pyards*pyard + tot_ryards*ryard + \
                 tot_ptd*ptd + tot_rtd*rtd - tot_tover*tover
    return (team,tot_points)
```

```python
team_list =["BUF","MIA","NE","NYJ","BAL","CIN","CLE","PIT","HOU","IND","JAC","TEN","DEN","KC","OAK","SD","DAL","NYG","PHI","WAS",
            "CHI","DET","GB","MIN","ATL","CAR","NO","TB","ARI","STL","SF","SEA"]
year_list = ["2010","2011","2012","2013","2014"]
```

```python
points_allowed_QB_dict={year:sorted([points_allowed(team,"QB",year) for team in team_list], \
                                 key=lambda x: x[1],reverse=True) for year in year_list}
```

```python
avg_ppg_allowed_QB= {year:[(tup[0],tup[1]/16.) for tup in points_allowed_QB_dict[year]]for year in year_list}          
```
